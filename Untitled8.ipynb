```python
from scipy.stats import f_oneway

age_groups = {}
for age, medv in zip(boston_df['AGE'], boston_df['MEDV']):
    if age not in age_groups:
        age_groups[age] = []
    age_groups[age].append(medv)

f_statistic, p_value = f_oneway(*age_groups.values())

print('F-Statistic:', f_statistic)
print('P-Value:', p_value)
```
